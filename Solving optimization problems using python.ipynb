Exercise 1

min -4x -2y

    s.t.
        x + y    <= 8

       8x + 3y  >= -24
       
      -6x + 8y  <= 48
      
       3x + 5y  <= 15

```python
import pyomo.environ as pyo
from pyomo.environ import *
from pyomo.opt import SolverFactory
import numpy as np
```

```python

model = pyo.ConcreteModel()

model.x = pyo.Var(bounds=(-np.inf, 3))
model.y = pyo.Var(bounds=(0, np.inf))

x = model.x
y = model.y

model.C1 = pyo.Constraint(expr= x + y <= 8)
model.C2 = pyo.Constraint(expr= 8*x + 3*y >= -24)
model.C3 = pyo.Constraint(expr= -6*x + 8*y <= 48)
model.C4 = pyo.Constraint(expr= 3*x + 5*y <= 15)

model.obj = pyo.Objective(expr= -4*x - 2*y, sense=minimize)

opt = SolverFactory('gurobi')

opt.solve(model)

model.pprint()

x_value = pyo.value(x)
y_value = pyo.value(y)

print("x =", {x_value})
print("y =", {y_value})


```

Example 2

The State College police department has the following minimum daily requirements for policemen on duty. There are 6 shifts, each shift requires 15,35,80,40,and 25 policemen, respectively and each shift is 4 hours-long. If The first shift start at 12 am and Each policeman comes on duty at 12am, 4am, 8am, 12pm, 4pm, or 8pm, and works for eight consecutive hours, find the number of policemen each shifts and total number of policemen.

```python
 # create an optimization model using the GLPK solver
required_policemen = [15, 35, 65, 80, 40, 25]
model = ConcreteModel()
    
model.num_policemen = Var(range(0, 6), domain=NonNegativeReals)

num_policemen = model.num_policemen
    
model.obj = Objective(expr=sum(num_policemen[i] for i in range(0, 6)), sense=minimize)
    
model.shift1 = Constraint(expr=num_policemen[0] + num_policemen[1] >= required_policemen[1])
model.shift2 = Constraint(expr=num_policemen[1] + num_policemen[2] >= required_policemen[2])
model.shift3 = Constraint(expr=num_policemen[2] + num_policemen[3] >= required_policemen[3])
model.shift4 = Constraint(expr=num_policemen[3] + num_policemen[4] >= required_policemen[4])
model.shift5 = Constraint(expr=num_policemen[4] + num_policemen[5] >= required_policemen[5])
model.shift6 = Constraint(expr=num_policemen[5] + num_policemen[0] >= required_policemen[0])
    
opt = SolverFactory('glpk')
opt.solve(model)
model.pprint()

print("Total of number of policemen", model.obj())

for i in range (0,6):

    print("Number of policemen for shift",i+1, "=",value(num_policemen[i]))


    
```
